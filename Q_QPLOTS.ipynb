# Q-Q Plot

```python
import numpy as np
import pylab
import scipy.stats as stats

#N(0,1)
std_normal = np.random.normal(loc=0, scale=1, size=1000)
# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
# Generate 100 samples from N(20,5)
measurements = np.random.normal(loc = 20, scale = 5, size = 100)
stats.probplot(measurements, dist="norm",plot=pylab)
pylab.show()
```

# if x & y lie on straight line then x & y have similar distribution 
# As my No. Of Samples (Size) increases, More & More points start lying on this line
# Limitaion: If No. of observation is small, its hard to interpret Q-Q Plot

```python
#generate 100 samples from N(20,5)
measurements = np.random.uniform(low=-1, high = 1, size = 100)
stats.probplot(measurements, dist="norm",plot=pylab)
pylab.show()
```

# Here order values is in uniform and Theoretical Quantiles is in Gaussian
# If most of the points are on staright line i.e., the distribution on X-axis & Y-axis are of same family
